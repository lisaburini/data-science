```python
# load dataset and show first rows 
# describe how many values it has -> 1 million rows
import pandas as pd
import numpy as np
pd.options.display.max_rows = 1000
full_dataset = pd.read_csv(r'card_transdata.csv')
display(full_dataset)
```

```python
# check null values -> no null values -> no management required
full_dataset.isnull().sum()
```

```python
# check types of attributes and change them if necessary
# describe types before and after etl
print(full_dataset.dtypes)

# repeat_retailer, ..., fraud are float: before cast them do binary check unique values in these columns
print("unique values for repeat_retailer: ", full_dataset["repeat_retailer"].unique())
print("unique values for used_chip: ", full_dataset["used_chip"].unique())
print("unique values for used_pin_number: ", full_dataset["used_pin_number"].unique())
print("unique values for online_order: ", full_dataset["online_order"].unique())
print("unique values for fraud: ", full_dataset["fraud"].unique())



# all these attributes have as unique values 1 and 0, so we can cast them to binary type
full_dataset['repeat_retailer'] = full_dataset['repeat_retailer'].apply(int)
full_dataset['used_chip'] = full_dataset['used_chip'].apply(int)
full_dataset['used_pin_number'] = full_dataset['used_pin_number'].apply(int)
full_dataset['online_order'] = full_dataset['online_order'].apply(int)
full_dataset['fraud'] = full_dataset['fraud'].apply(int)

print(full_dataset.dtypes)

# check 0 and 1, no more 0. and 1.
print("unique values for repeat_retailer: ", full_dataset["repeat_retailer"].unique())
print("unique values for used_chip: ", full_dataset["used_chip"].unique())
print("unique values for used_pin_number: ", full_dataset["used_pin_number"].unique())
print("unique values for online_order: ", full_dataset["online_order"].unique())
print("unique values for fraud: ", full_dataset["fraud"].unique())
```

```python
# final dataset after preprocessing
display(full_dataset)

# export dataset in csv
full_dataset.to_csv('card_transdata_preprocessed.csv', index=False)
```
